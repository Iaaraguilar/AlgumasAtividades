```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re 
df = pd.read_excel('financeiro.xlsx', sheet_name='Despesas')
```

```python
df['Data']=pd.to_datetime(df['Data'],unit='ms')
df['Data']=df['Data'].dt.strftime('%d/%m/%Y')
df
```

```python
df['Fornecedor'].fillna('Desconhecido',inplace=True)
df.head(20)
```

```python
df['Valor'].fillna('Desconhecido',inplace=True)
df.head(20)
```

```python

def remover_siglas(nome):
    siglas = ['S/A', 'EI', 'LTDA', 'ME', 'EIRELI', 'Ltda', 'SA', 'SA.']
    pattern = r'\b(?:' + '|'.join(siglas) + r')\.?$'
    return re.sub(r'\s*' + pattern, '', nome, flags=re.IGNORECASE).strip()


df['Fornecedor'] = df['Fornecedor'].apply(remover_siglas)

df['Fornecedor'] = df['Fornecedor'].str.replace(r'\s*-\s*', '', regex=True).str.strip()

df.head(20)
```

```python

df = df[~df.apply(lambda row: row.astype(str).str.contains('desconhecido', case=False).any(), axis=1)]
df.head(20)
```

```python
import pandas as pd

# Suponha que você tenha esse DataFrame:
df_orcamentos = pd.DataFrame({
    'Mes': list(range(1, 13)),
    'Orcamento': [1000, 1100, 1200, 1300, 1250, 1400, 1500, 1600, 1550, 1650, 1700, 1800]
})

# Dicionário para substituir número do mês por abreviação
mesesd = {
    1: 'jan', 2: 'fev', 3: 'mar', 4: 'abr',
    5: 'mai', 6: 'jun', 7: 'jul', 8: 'ago',
    9: 'set', 10: 'out', 11: 'nov', 12: 'dez'
}

# Cria uma nova coluna com os nomes dos meses
df_orcamentos['Nome_mes'] = df_orcamentos['Mes'].replace(mesesd)
df_orcamentos
```
